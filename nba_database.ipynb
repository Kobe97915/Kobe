```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("NBA_2024_cleaned.csv")

# Define high scorers (PTS >= 20)
high_scorers = df[df["PTS"] >= 20]
other_players = df[df["PTS"] < 20]

# Metrics to compare
metrics = ["FGA", "3PA", "FT", "AST", "TRB"]

# Calculate mean values for both groups
high_scorers_mean = high_scorers[metrics].mean()
other_players_mean = other_players[metrics].mean()

# Create a comparison DataFrame
comparison_df = pd.DataFrame({"High Scorers": high_scorers_mean, "Other Players": other_players_mean})

# Plot the data
plt.figure(figsize=(10, 6))
comparison_df.plot(kind="bar", width=0.8, figsize=(10, 6), color=["blue", "lightgray"])

# Customize plot
plt.title("Comparison of Key Metrics Between High Scorers and Other Players", fontsize=14)
plt.ylabel("Mean Value", fontsize=12)
plt.xticks(rotation=0)
plt.legend(title="Player Group", loc="upper right")

# Show the plot
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```
